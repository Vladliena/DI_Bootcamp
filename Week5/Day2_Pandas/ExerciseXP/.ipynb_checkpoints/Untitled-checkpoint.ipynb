```python
figure,axes = plt.subplots()
axes.set(title='Line plot', xlabel='X',ylabel='Y')
axes.plot([1,2,3,4],[2,4,5,7], color='green', linestyle = '--')
```

```python
figure,axes = plt.subplots()
axes.set(title='Line plot', xlabel='X',ylabel='Y')
axes.plot([1,2,3,4],[2,4,5,7], color='green', linestyle = '--', marker = 'o', mfc='red',mec='blue', label = 'Yossi')
```

```python
figure,axes = plt.subplots()
axes.set(title='Line plot', xlabel='X',ylabel='Y')
axes.plot([1,2,3,4],[2,4,5,7], color='green', linestyle = '--', marker = 'o', mfc='red',mec='blue', label = 'Yossi')
axes.plot([5,6,7,8],[4,5,7,9], color='pink', linestyle = '--', marker = 'o', mfc='red',mec='blue', label = 'Ziv')
axes.legend()
figure.savefig('my_plot.png')
```

```python
x = np.linspace(0, 10, 20)
figure,axes = plt.subplots()
y1 = x ** 2
y2 = x ** 1.5
axes.scatter(x,y1,y2)
figure.savefig('my_plot1.png')
```

```python
x = np.linspace(0, 10, 20)
figure,axes = plt.subplots()
y1 = x ** 2
y2 = x ** 1.5
axes.plot(x,y1,color='red', linestyle = 'dashed')
axes.set(title='y = x^2')
```

```python
x = np.linspace(0, 10, 20)
figure,axes = plt.subplots()
y1 = x ** 2
y2 = x ** 1.5
axes.plot(x,y1,color='red', linestyle = 'dotted', marker='s',label='Y')
axes.plot(x,y2,color='red', linestyle = 'dotted', marker='s',label='Y')
axes.legend()
axes.set(title='y = x^2')
```

```python
x = np.linspace(1,100,1000)
y1 = x **2
y2 = x ** 3
figure, axes_set = plt.subplots(2)

axes_set[0].plot(x,y1, color = 'red')
axes_set[1].plot(x,y2, color='green')
```

```python
x = np.linspace(1,100,1000)
y1 = x **2
y2 = x ** 3
figure, axes_set = plt.subplots(2)

axes_set[0].plot(x,y1, color = 'red')
axes_set[0].annotate('breakout!', xy=(0,0), xytext=(1,2000), arrowprops={'facecolor': 'black','shrink':10})
axes_set[1].plot(x,y2, color='green')

plt.show()
```

```python
x = np.linspace(1,100,1000)
y1 = x **2
y2 = x ** 3
figure, axes_set = plt.subplots(2)

axes_set[0].plot(x,y1, color = 'red', label='main 1 line')
axes_set[0].legend(title='Legend', loc='lower right', shadow=True)
axes_set[0].annotate('breakout!', xy=(0,0), xytext=(1,2000), arrowprops={'facecolor': 'black','shrink':10})
axes_set[1].plot(x,y2, color='green', label = 'main 2 line')
axes_set[1].legend()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
figure,axes_array = plt.subplots(4)
axes_array[0].plot(x,y)
axes_array[1].plot(x,-y)
axes_array[2].plot(x,y**2)
axes_array[3].plot(x, -y**2)
```
