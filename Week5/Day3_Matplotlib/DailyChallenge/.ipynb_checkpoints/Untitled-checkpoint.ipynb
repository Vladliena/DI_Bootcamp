```python
# 1
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Task 1

```python
# 2
x = np.linspace(0, 20, 100)
# 3
y = x ** 2
# 4
plt.plot(x,y)
# 5
plt.title('Plot of x vs y (y = x^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,20)
plt.ylim(0,400)
# 6
plt.savefig('plot_of_x_vs_y.png')
```

# Task 2

```python
# 1
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')
```

```python
#2
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
```

```python
figure,axes = plt.subplots()
plt.scatter(chipo['item_price'], chipo['quantity'])
# 3
axes.set(title='Relationship between item_prices and quantity')
# 4
plt.xlabel('Item Price')
plt.ylabel('Quantity')
# 5
plt.savefig('items_price_and_quantity_relationship.png')
```

# Task 3

```python
#1
item_totals = chipo.groupby('item_name')['quantity'].sum().nlargest(5)
```

```python
item_totals
```

```python
#2 / # 3
figure,axes = plt.subplots()
axes.set(title = 'Top 5 Items by quantity ordered')
item_totals.plot(kind='bar')
plt.xlabel('Item Name')
plt.ylabel('Quantity Ordered')
# 4
plt.savefig('top_5_items.png')
```
