```python
import numpy as np
```

```python
# Exercise 7 / Exercise 8

x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
figure,axes_array = plt.subplots(2)
axes_array[0].plot(x,y_sin, color='tab:blue')
axes_array[1].plot(x,y_cos, color='tab:blue')
axes_array[0].set_title('y = sin(x)')
axes_array[1].set_title('y = cos(x)')

axes_array[0].legend()
axes_array[1].legend()
```

```python
# Exercise 7 / Exercise 8

x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
figure,axes_array = plt.subplots(2)
axes_array[0].scatter(x,y_sin, color='tab:blue', marker='X')
axes_array[1].scatter(x,y_cos, color='tab:blue', marker='X')
axes_array[0].set_title('y = sin(x)')
axes_array[1].set_title('y = cos(x)')

axes_array[0].legend()
axes_array[1].legend()
```

```python
# Exercise 9
x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
figure,axes_array = plt.subplots(2)
axes_array[0].scatter(x,y_sin, label='y = sin(x)', color='tab:blue', marker='o')
axes_array[1].scatter(x,y_cos, label='y = cos(x)',color='tab:blue', marker='X')

axes_array[0].legend()
axes_array[1].legend()
```
